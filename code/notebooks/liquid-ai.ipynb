```python
def gen_document(document_metadata):
    from openai import OpenAI
    model = "/models/LiquidCloud"

    API_KEY= 'secret_testingreport-grafana_88c10e40213144f58eb24acb9fa1a1c3.IEDFHewBrwVx7ccd2SUshQcMXb3bqccK'

    openai_api_key = API_KEY
    openai_api_base = "https://api.lambdalabs.com/v1"

    client = OpenAI(
        api_key=openai_api_key,
        base_url=openai_api_base
    )
    # client.models.list()
    # # iterate over the list
    # for model in client.models.list():
    #     print(model.id)

    chat_completion = client.chat.completions.create(
        messages=[{
            "role": "system",
            "content": "You are a helpful assistant made by to generate a hypothetical coorporate document for a coffee shop based on the given metadata."
        }, {
            "role": "user",
            "content": f"metadata: {str(document_metadata)}"
        }],
        model=model,
        temperature=0.7
    )

    return chat_completion.choices[0].message.content
```
